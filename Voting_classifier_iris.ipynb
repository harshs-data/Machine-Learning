```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
original_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
original_df['target'] = iris.target

df = original_df[original_df['target'] != 0][["sepal length (cm)","sepal width (cm)", "target" ]]
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:, : 2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
clf1 = LogisticRegression()
clf2 = KNeighborsClassifier()
clf3 = RandomForestClassifier()
```

```python
estimators = [('lr', clf1), ('knn', clf2), ('rf', clf3)]
```

```python
for estimator in estimators:
  x = cross_val_score(estimator[1],X,y,cv=10,scoring='accuracy')
  print(estimator[0],np.round(np.mean(x),2))
```

```python
from sklearn.ensemble import VotingClassifier
```

### Hard voting

```python
vc = VotingClassifier(estimators=estimators, voting='hard')
x = cross_val_score(vc,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))
```

### Soft voting


```python
vc1 = VotingClassifier(estimators=estimators, voting='soft')
x = cross_val_score(vc1,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))
```

### Weighted voting


```python
for i in range(1, 4):
    for j in range(1, 4):
        for k in range(1, 4):
            # Define the Voting Classifier with 'soft' voting
            vc = VotingClassifier(estimators=estimators, voting='soft', weights=[i, j, k])

            # Perform 10-fold cross-validation using the scaled data
            x = cross_val_score(vc, X, y, cv=10, scoring='accuracy')

            # Print the weights and the mean accuracy rounded to 2 decimal places
            print("for i={}, j={}, k={}: {:.2f}".format(i, j, k, np.round(np.mean(x), 2)))
```

### Classifier of same algo

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

# 1. Generate the classification dataset
X, y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=2)

# 2. Define SVM models with different polynomial degrees
# Note: 'degree' controls the complexity/curviness of the boundary
svm1 = SVC(probability=True, kernel='poly', degree=1)
svm2 = SVC(probability=True, kernel='poly', degree=2)
svm3 = SVC(probability=True, kernel='poly', degree=3)
svm4 = SVC(probability=True, kernel='poly', degree=4)
svm5 = SVC(probability=True, kernel='poly', degree=5)

# 3. Create the list of estimators (model-name pairs)
estimators = [('svm1', svm1), ('svm2', svm2), ('svm3', svm3), ('svm4', svm4), ('svm5', svm5)]

# 4. Loop through each estimator to perform 10-fold cross-validation
# 'estimator' is the singular temporary variable for each pair in 'estimators'
for estimator in estimators:
    # Use cross_val_score to get 10 accuracy results for the model at index [1]
    x = cross_val_score(estimator[1], X, y, cv=10, scoring='accuracy')

    # Print the name at index [0] and the average of the 10 scores
    print(estimator[0], np.round(np.mean(x), 2))
```

```python
vc1 = VotingClassifier(estimators=estimators, voting='soft')
x = cross_val_score(vc1, X, y, cv=10, scoring='accuracy')
print(np.round(np.mean(x), 2))
```
